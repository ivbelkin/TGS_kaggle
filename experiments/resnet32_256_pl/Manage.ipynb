```python
def plot(df, name):
    N = df.shape[0]
    plt.figure(figsize=[15, 7])
    
    plt.plot(np.arange(1, N + 1), df["train_" + name], "b")
    plt.plot(np.arange(1, N + 1), df["valid_" + name], "r")
    
    plt.grid()
```

```python
plot(df, "mAP_0.5")
#plt.ylim(0.7, 1.0)
```
